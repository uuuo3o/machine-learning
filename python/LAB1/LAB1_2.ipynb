```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
# m1 = [10, 5]
m1 = [20, 7]
v1 = [[25, 0], [0, 9]]
x1 = np.random.multivariate_normal(m1, v1, N)

# m2 = [10, 5]
m2 = [20, 7]
v2= [[25, 10], [10, 9]]
x2 = np.random.multivariate_normal(m2, v2, N)

# m3 = [10, 5]
m3 = [20, 7]
v3 = [[25, -10], [-10, 9]]
x3 = np.random.multivariate_normal(m3, v3, N)
```

```python
plt.scatter(x1[:, 0], x1[:, 1], s=10, color='b')
plt.scatter(x2[:, 0], x2[:, 1], s=10, color='g')
plt.scatter(x3[:, 0], x3[:, 1], s=10, color='r')
```
